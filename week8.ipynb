### 3. Buatlah fungsi dilasi serta tampilkan hasil dilasi pada citra dengan Structuring Element Dilasi (SED) berbentuk cross 3 x 3 tanpa menggunakan library morphology dari openCV, sehingga menghasilkan tampilan seperti di bawah ini:

```python
#Dilasi tanpa library

def dilasi_citra(F,w):
  #size image
  # p : widith
  # q : height
  # F : untuk mengetahui lebar dan tinggi citra
  p,q= F.shape

  # inisialisasi dengan nilai nol dengan ukuran (height dan width) sesuai dengan parameter F (citra gambar)
  imgD= np.zeros((p,q), dtype=np.uint8)

  #Generate structure element dilasi (SED) seperti kernel
  SED= np.ones((w,w), dtype=np.uint8)
  # print("SED : ",SED)
  constant1= (w-1)//2
  # print("constant1", constant1)

  #Proses Dilasi
  for i in range(constant1, p-constant1):
    for j in range(constant1,q-constant1):
      # F[baris, kolom]
      # F[range baris, range kolom]
      # F[0 : 10, 5 : 20] = berisi baris 0 sampai 10 dan kolom 5 sampai 20

      # temp matrix 3 x 3
      temp= F[i-constant1:i+constant1+1, j-constant1:j+constant1+1]

      # print("temp =============")
      # print(temp)
      # print("=============")

      # proses perkalian temp*SED
      # cont1 = [[0, 255, 255],
      #    [0, 0, 255],
      #    [255, 0, 0]]

      # SED = [[1, 1, 1],
      #  [1, 1, 1],
      #  [1, 1, 1]]

      # proses = [[0*1, 49*1, 134*1],
      #      [109*1, 49*1, 139*1],
      #      [200*1, 9*1, 134*1]]
      product= temp*SED

      # mengambil nilai pixel maximum dari matrix produuk
      imgD[i,j]= np.max(product)
  # return imgD, (p,q)
  # return imgD, (constant1)
  # return imgD,(temp)
  return imgD

img_d= cv2.imread('/content/drive/MyDrive/content/Images/platnomor.png',0)
# cv2_imshow(img_d)

# Menerapkan thresholding pada citra img_d dengan nilai threshold 125,
# mengubah piksel yang lebih besar dari threshold menjadi 200, dan
# menyimpan hasilnya dalam variabel thresh.
# cv2.threshold = mengubah citra menjadi citra biner (hitam putih)
# nilai pixel  kurang dari 125 diklasifikasikan sebagai hitam (0)
# nilai pixel lebih dari 125 menjadi putih (200)
# pemilihan nilai threshold() bergantung pada citra (pencahayaan, tingkat noise, dan dll)
ret, thresh = cv2.threshold(img_d, 125, 200, cv2.THRESH_BINARY)
w=3

# imgD,(p,q) = dilasi_citra(thresh,w)
# imgD, (constant1) = dilasi_citra(thresh,w)
# imgD, temp = dilasi_citra(thresh,w)
imgD= dilasi_citra(thresh,w)
# print(thresh)
# print(ret)


# print('nilai p : ' , p)
# print('nilai q : ', q)
# print('constant1 : ', constant1)
# print("temp : ", temp)


plt.subplot(131),plt.imshow(img_d,cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(imgD,cmap = 'gray')
plt.title('Citra Hasil Dilasi'), plt.xticks([]), plt.yticks([])
plt.show()
```

### 5. Buatlah fungsi erosi serta tampilkan hasil erosi pada citra dengan Structuring Element Erosi (SE) berbentuk square 5 x 5 tanpa menggunakan library morphology dari openCV, sehingga menghasilkan tampilan seperti di bawah ini:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Erosi Tanpa Library
def erosi_citra(F, k):
    # Size image
    m, n = F.shape

    # Generate structure element erosi (SE)
    SE = np.ones((k, k), dtype=np.uint8)
    constant = (k - 1) // 2
    imgE = np.zeros((m, n), dtype=np.uint8)

    # Proses Erosi
    for i in range(constant, m - constant):
        for j in range(constant, n - constant):
            temp = F[i - constant:i + constant + 1, j - constant:j + constant + 1]
            product = temp * SE

            # mencari nilai terkecil dari array product
            imgE[i, j] = np.min(product)

    return imgE

img_e = cv2.imread('/content/drive/MyDrive/content/Images/platnomor.png', 0)
ret1, thresh1 = cv2.threshold(img_e, 175, 225, cv2.THRESH_BINARY)
k = 5
imgE = erosi_citra(thresh1, k)

# plt.subplot(131) : 1 baris ,3 kolom , kolom pertama
# plt.xtricks([]) : menghapus kordinat x
# plt.ytricks([]) : menghapus kordinat y
# cmap="gray" : mengatur warna menjadi gray
plt.subplot(131), plt.imshow(img_e, cmap='gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(imgE, cmap='gray')
plt.title('Citra Hasil Erosi'), plt.xticks([]), plt.yticks([])
# plt.subplot(133) ,plt.imshow(thresh1,cmap="gray")
# plt.title('Citra Biner'), plt.xticks([]), plt.yticks([])
plt.show()
```

### 7. Buatlah program implementasi operasi opening dengan Structuring Element berbentuk square 3 x 3 tanpa dan dengan menggunakan library morphology dari openCV, sehingga menghasilkan tampilan seperti di bawah ini: Operasi Opening Menggunakan tanpa Library Morphology

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# opening : erosi - dilasi
# Erosi Tanpa Library
def erosi_citra(F, k):
    # Size image
    m, n = F.shape

    # Generate structure element erosi (SE)
    SE = np.ones((k, k), dtype=np.uint8)
    constant = (k - 1) // 2
    imgE = np.zeros((m, n), dtype=np.uint8)

    # Proses Erosi
    for i in range(constant, m - constant):
        for j in range(constant, n - constant):
            temp = F[i - constant:i + constant + 1, j - constant:j + constant + 1]
            product = temp * SE
            imgE[i, j] = np.min(product)

    return imgE

# Dilasi Tanpa Library
def dilasi_citra(F, k):
    # Size image
    p, q = F.shape

    imgD = np.zeros((p, q), dtype=np.uint8)

    # Generate structure element dilasi (SED)
    SED = np.ones((k, k), dtype=np.uint8)
    constant1 = (k - 1) // 2

    # Proses Dilasi
    for i in range(constant1, p - constant1):
        for j in range(constant1, q - constant1):
            temp = F[i - constant1:i + constant1 + 1, j - constant1:j + constant1 + 1]
            product = temp * SED
            imgD[i, j] = np.max(product)

    return imgD

img_o = cv2.imread('/content/drive/MyDrive/content/Images/platnomor.png', 0)
ret, thresho = cv2.threshold(img_o, 127, 225, cv2.THRESH_BINARY)
k = 7
imgO = dilasi_citra(erosi_citra(thresho, k), k)

plt.subplot(131), plt.imshow(img_o, cmap='gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(imgO, cmap='gray')
plt.title('Citra Hasil Opening'), plt.xticks([]), plt.yticks([])
plt.show()
```

### 8. Buatlah program implementasi operasi closing dengan Structuring Element berbentuk square 3 x 3 tanpa dan dengan menggunakan library morphology dari openCV, sehingga menghasilkan tampilan seperti di bawah ini: operasi closing tanpa library morphology

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# closing : dilasi - erosi
# Erosi Tanpa Library
def erosi_citra(F, k):
    m, n = F.shape
    SE = np.ones((k, k), dtype=np.uint8)
    constant = (k - 1) // 2
    imgE = np.zeros((m, n), dtype=np.uint8)

    for i in range(constant, m - constant):
        for j in range(constant, n - constant):
            temp = F[i - constant:i + constant + 1, j - constant:j + constant + 1]
            product = temp * SE
            imgE[i, j] = np.min(product)

    return imgE

# Dilasi Tanpa Library
def dilasi_citra(F, w):
    p, q = F.shape
    imgD = np.zeros((p, q), dtype=np.uint8)
    SED = np.ones((w, w), dtype=np.uint8)
    constant1 = (w - 1) // 2

    for i in range(constant1, p - constant1):
        for j in range(constant1, q - constant1):
            temp = F[i - constant1:i + constant1 + 1, j - constant1:j + constant1 + 1]
            product = temp * SED
            imgD[i, j] = np.max(product)

    return imgD

img_path = '/content/drive/MyDrive/content/Images/platnomor.png'
img_c = cv2.imread(img_path, 0)

if img_c is not None:
    retc, threshc = cv2.threshold(img_c, 127, 225, cv2.THRESH_BINARY)
    k = 7
    imgDilated = dilasi_citra(threshc, k)
    imgClosed = erosi_citra(imgDilated, k)

    plt.subplot(131), plt.imshow(img_c, cmap='gray')
    plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
    plt.subplot(132), plt.imshow(imgClosed, cmap='gray')
    plt.title('Citra Hasil Closing'), plt.xticks([]), plt.yticks([])
    plt.show()
else:
    print("Citra tidak dapat ditemukan atau tidak terbaca dengan benar.")
```

```python
import cv2 as cv
from matplotlib import pyplot as plt
from skimage import filters
from skimage.morphology import skeletonize

# Baca citra
img = cv.imread('/content/drive/MyDrive/content/Images/lily.jpg', 0)

# Konversi citra menjadi citra biner
binary = img > filters.threshold_triangle(img)
# filters.threshold_triangle(img) : 253
# binary = membangdingkan seluruh nilai pixel dengan nilai 253

# Konversi nilai True/False menjadi 1/0
binary_cp = binary.astype(int)

# Skeletonize citra biner
skeleton = skeletonize(binary_cp)

# Tampilkan citra-citra
plt.figure(figsize=(20, 20))

plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(skeleton, cmap='gray')
plt.title('Skeleton Image'), plt.xticks([]), plt.yticks([])

plt.savefig("binary_and_bin_skeleton.png")
plt.show()
```

### 16. Skeleton Inverse

```python
from skimage import io
from matplotlib import pyplot as plt
from skimage import filters
from skimage.morphology import skeletonize

# Baca citra
img = io.imread('/content/drive/MyDrive/content/Images/lily.jpg', as_gray=True)

# Ubah citra menjadi citra biner menggunakan threshold Triangle
binary = img < filters.threshold_triangle(img)

# Skeletonize image
skeleton = skeletonize(binary)

# Tampilkan citra asli dan hasil skeletonization
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img, cmap='gray')
ax[0].set_title('Citra Asli')
ax[0].set_axis_off()

# Perhatikan colormap yang dibalik (gray_r) untuk citra skeleton
ax[1].imshow(skeleton, cmap='gray_r')
ax[1].set_title('Skeleton (Terbalik)')
ax[1].set_axis_off()

plt.show()
```
